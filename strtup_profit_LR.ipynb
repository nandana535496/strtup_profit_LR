```python
#objective:  build a linear regression model to estimate the expected profit, and explore which features contribute most to the business success.
import pandas as pd 
```

```python
#load the data_set
df=pd.read_csv("Downloads/archive (2).zip")
df.head()
```

```python
df.info()
```

```python
# One-Hot Encode the 'State' column
df = pd.get_dummies(df, columns=['State'], drop_first=True)
```

```python
df
```

```python
x=df.drop(['Profit'],axis=1)
y=df.Profit
```

```python
#Train/Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Build the Linear Regression Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#prediction
y_pred=lr.predict(X_test)
print(y_pred)
```

```python
# Evaluate the Model
from sklearn.metrics import mean_squared_error,r2_score
print(f'RMSE{mean_squared_error(y_test,y_pred)}')
print(f'r2_score{r2_score(y_test,y_pred)}')
```

```python
coeff_df = pd.DataFrame(lr.coef_, x.columns, columns=['Coefficient'])
print(coeff_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'red')

plt.xlabel("Actual Profit")
plt.ylabel("Predicted Profit")
plt.title("Actual vs Predicted Profit")
plt.grid(True)
plt.show()
```

```python
residuals = y_test - y_pred
plt.figure(figsize=(8, 4))
sns.histplot(residuals,  bins=30, color="purple")
plt.title("Distribution of Residuals")
plt.xlabel("Residual")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x=df["R&D Spend"], y=df["Profit"], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Profit vs R&D Spend")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
corr = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
